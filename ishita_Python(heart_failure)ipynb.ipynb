# Feature engineering

```python
# Define a function to categorize age into "Teen," "Adult," or "Aged"
def Frequency_of_Age(values):
    if values <= 20:
        return "Teen"
    elif values <= 40:
        return "Adult"
    else:
        return "Aged"
```

```python
# Create a new column 'DEATH_EVENT' in DataFrame 'df' by applying the function 'Frequency_of_Age' to the 'age' column
df['DEATH_EVENT'] =df.age.apply(Frequency_of_Age)
```

```python
# Display the first few rows of the DataFrame (df)
df.head()
```

# ***Encoding categorical***
# label encoder

```python
# Convert object-type columns to category type and encode them as codes
for diabetes in df.columns:
    if df[diabetes].dtype == 'object':
        df[diabetes] = pd.to_numeric(df[diabetes], errors='coerce')
        df[diabetes + '_codes'] = df[diabetes].astype('category').cat.codes
```

# **t**-**test**

```python
# Import the t-test function from the scipy.stats module
from scipy import stats

# Perform an independent two-sample t-test for 'diabetes' based on 'anaemia' values
t_statistics, p_value = stats.ttest_ind(df[df['anaemia'] == 1]['diabetes'].dropna(), df[df['anaemia'] == 0]['diabetes'].dropna())

# Print the calculated t-statistics
print(t_statistics)

# Print the corresponding p-value
print(p_value)
 # null hypothesis will be accepted
```

```python
# Import the required function from statsmodels
from statsmodels.stats.weightstats import ztest

# Perform a Z-test on the 'DEATH_EVENT' column
z_statistic, p_value = ztest(df['DEATH_EVENT'].dropna(), value=32)

# Print the results of the Z-test
print("\nZ-Test:")
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)
```

```python
# Perform one-way ANOVA for the 'DEATH_EVENT' column based on different education levels
anova= stats.f_oneway(df[df['smoking']==1]['DEATH_EVENT'].dropna(),
                      df[df['smoking']==2]['DEATH_EVENT'].dropna(),
                      df[df['smoking']==3]['DEATH_EVENT'].dropna(),
                      df[df['smoking']==4]['DEATH_EVENT'].dropna())
# The ANOVA test statistic
anova.statistic
```

```python
# Access the p-value from the ANOVA results
anova_pvalue = anova.pvalue
# The 'anova_pvalue' variable now contains the p-value obtained from the ANOVA test
```

# **Interpreting key findings from the analysis of the "Heart Failure Clinical Records" dataset:**

**Age Distribution:**

The dataset comprises individuals with varying ages, providing a representative sample of the population under consideration.

**Prevalence of Anaemia:**

The 'anaemia' column indicates the presence (1) or absence (0) of anaemia in patients.
The analysis may reveal the overall prevalence and distribution of anaemia in the dataset.

**Creatinine Phosphokinase Levels:**

The 'creatinine_phosphokinase' column represents enzyme levels, and its distribution can provide insights into potential trends or outliers.

**Diabetes Occurrence:**

The 'diabetes' column reflects whether a patient has diabetes (1) or not (0).
Understanding the frequency and distribution of diabetes in the dataset is crucial for risk assessment.

**Ejection Fraction Insights:**

The 'ejection_fraction' column indicates the percentage of blood leaving the heart during contraction.
Examining this variable may offer insights into cardiac function and potential correlations with other features.

**High Blood Pressure Prevalence:**

The 'high_blood_pressure' column denotes the presence (1) or absence (0) of high blood pressure.
Understanding the prevalence and impact of high blood pressure is vital for cardiovascular health analysis.

**Platelet Counts:**

The 'platelets' column signifies platelet levels in the blood, and its distribution may indicate potential abnormalities or trends.

**Serum Creatinine and Sodium Levels:**

The 'serum_creatinine' and 'serum_sodium' columns provide information on kidney function and electrolyte balance, respectively.
Analyzing these variables can aid in assessing overall renal health.

**Gender Distribution:**

The 'sex' column denotes the gender of individuals (0 for female, 1 for male).
Understanding the gender distribution is crucial for identifying potential gender-specific health patterns.

**Smoking Habits and Time Duration:**

The 'smoking' column indicates whether a patient is a smoker (1) or non-smoker (0).
The 'time' column represents the follow-up period for each patient.
Examining smoking habits and follow-up duration can offer insights into lifestyle factors and their impact on health outcomes.

**Mortality Rates (DEATH_EVENT):**

The 'DEATH_EVENT' column indicates whether a patient experienced a death event (1) or not (0).
